# Импортируем нужные модули

```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.utils import shuffle
from matplotlib import style
```

# Анализируем данные

```python
data = pd.read_csv("/content/drive/MyDrive/DataScience/StudentPerformanceLR/student-mat.csv", sep=";")

print(data.head())
```

```python
data = data[["G1", "G2", "G3", "studytime", "failures", "absences", "Medu","Fedu"]]
print(data.head())
```

# Разбиваем данные на выборки

```python
predict = "G3"
```

```python
x = np.array(data.drop([predict],1))
y = np.array(data[predict])
```

```python
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x,y, test_size = 0.1)
```

# Создаем модель линейной регрессии и обучаем ее

```python
linear = linear_model.LinearRegression()
```

```python
linear.fit(x_train, y_train)
```

```python
acc = linear.score(x_test,y_test)
print(acc)
```

# Выводим прогнозы и тестовые значения

```python
predictions = linear.predict(x_test)

for x in range(len(predictions)):
  print(round(predictions[x], 1),"----", y_test[x], x_test[x])
```
